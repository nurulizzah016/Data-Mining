## Data Preprocessing

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('ShenyangPM.csv',sep=';')
data.head()
```

```python
np.sum(data.isnull())
```

```python
data.describe()
```

```python
data['PM_Taiyuanjie']=data['PM_Taiyuanjie'].fillna((data['PM_Taiyuanjie'].mean()))
data['PM_US Post']=data['PM_US Post'].fillna((data['PM_US Post'].mean()))
data['PM_Xiaoheyan']=data['PM_Xiaoheyan'].fillna((data['PM_Xiaoheyan'].mean()))
data['DEWP']=data['DEWP'].fillna((data['DEWP'].mean()))
data['HUMI']=data['HUMI'].fillna((data['HUMI'].mean()))
data['PRES']=data['PRES'].fillna((data['PRES'].mean()))
data['TEMP']=data['TEMP'].fillna((data['TEMP'].mean()))
data['cbwd']=data['cbwd'].fillna(data['cbwd'].dropna().mode()[0])
data['Iws']=data['Iws'].fillna((data['Iws'].mean()))
data['precipitation']=data['precipitation'].fillna((data['precipitation'].mean()))
data['Iprec']=data['Iprec'].fillna((data['Iprec'].mean()))
np.sum(data.isnull())
```

### Mengatasi Outlier

```python
from scipy import stats
data1=data.drop('cbwd',1)
data2=data.drop('No',1)
z=np.abs(stats.zscore(data2._get_numeric_data()))
print(z)
data=data[(z<3).all(axis=1)]
print(data.shape)
```
