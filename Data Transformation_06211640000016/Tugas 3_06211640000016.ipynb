## Missing Value

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('class grade.csv',sep=';')
data=data.drop('Prefix',1)
data.tail()
```

```python
np.sum(data.isnull())
```

```python
data['TakeHome']=data['TakeHome'].fillna((data['TakeHome'].mean()))
data['Final']=data['Final'].fillna((data['Final'].mean()))
np.sum(data.isnull())
```

```python
data.to_csv('D:\ITS\semester 6\Data Mining\class grade_1.csv')
data1=data
data1.head()
print(data1.shape)
```

## Missing Value & Outlier

```python
from scipy import stats
z=np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data2=data[(z<3).all(axis=1)]
print(data2.shape)
```

```python
data2.to_csv('D:\ITS\semester 6\Data Mining\class grade_2.csv')
```

## Missing Value, Transformation 

```python
x=data.iloc[0:99,0:5]
x.head()
```

```python
from sklearn.preprocessing import normalize
xnorm=normalize(x,norm='l2') 
xnorm=pd.DataFrame(xnorm) 
xnorm.tail()
```

```python
xnorm.to_csv('D:\ITS\semester 6\Data Mining\class grade_3.csv')
print(xnorm.shape)
```
