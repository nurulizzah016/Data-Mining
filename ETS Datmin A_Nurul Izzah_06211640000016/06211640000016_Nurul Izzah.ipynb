# Pre Processing Data

```python
data.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
data['InvoiceNo']=data['InvoiceNo'].astype('str')
data = data [~ data['InvoiceNo'].str.contains('C')]
data['UnitPrice']=data['UnitPrice'].astype('str')
data = data [~ data['UnitPrice'].str.contains('0')]
np.sum(data.isnull())
```

```python
#mengelompokkan barang-barang menjadi satu per transaksi berdasarkan InvoiceNo
x = data.groupby('InvoiceNo')
transaksi = []
for i,j in x:
    transaksi.append(list(j['Description'].map(str)))
```

```python
x=pd.DataFrame(transaksi)
x.head(n=6)
```

```python
x.shape
```

## Analisis Association Rule untuk semua Negara

```python
records=[]
for i in range (0, 19168):
    records.append([str(x.values[i,j]) for j in range (0,923)])
```

```python
association_rules=apriori(records,min_support=0.008,
                         min_confidence=0.4,min_lift=6)
association_result=list(association_rules)
association_result[0]
```

```python
for item in association_result:
#first index of the inner list
#contains base item and add item
    pair = item[0]
    items = [y for y in pair]
    print("rule: " + items [0] + " ->" + items [1])
    
    #second index of the inner list
    print("support: " + str(item[1]))
    
    #third index of the list located at 10th
    #of the third index of the inner list
    
    print("confidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("=====================================")
```

## Analisis Association Rule untuk Negara Netherlands

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
data1=pd.read_csv('online_retail.csv')
```

```python
data1['Description']=data1['Description'].str.strip()
data1.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
data1['InvoiceNo']=data1['InvoiceNo'].astype('str')
data1=data1[~data1['InvoiceNo'].str.contains('C')]
data1.head(n=5)
```

```python
data1['Country'].unique()
```

## Basket Analysis for Netherlands

```python
basket=(data1[data1['Country']=='Netherlands']
        .groupby(['InvoiceNo','Description'])['Quantity']
        .sum().unstack().reset_index().fillna(0)
        .set_index('InvoiceNo'))
basket.head(n=10)
```

```python
basket.shape
```

```python
def encode_units(x):
    if x <=0:
        return 0
    if x >=1:
        return 1
    
basket_sets=basket.applymap(encode_units)
basket_sets.drop('POSTAGE',inplace=True,axis=1)
```

```python
frequent_itemset=apriori(basket_sets,min_support=0.07,use_colnames=True)
rules=association_rules(frequent_itemset,metric="lift",min_threshold=1)
```

```python
z=rules[(rules['lift']>=8)&
       (rules['confidence']>=0.7)]
rules1=z.drop(columns=['antecedent support','consequent support','leverage','conviction'])
rules1
```
