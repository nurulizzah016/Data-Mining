```python
data.tail()
```

```python
from scipy import stats
data1=data.drop(['User_ID','Product_ID','Gender','Age','City_Category','Stay_In_Current_City_Years','Marital_Status'],1)
z=np.abs(stats.zscore(data1._get_numeric_data()))
print(z)
data2=data[(z<3).all(axis=1)]
print(data2.shape)
```

```python
y=data2.describe()
y.to_csv('D:\ITS\semester 6\Data Mining1\Statdes2.csv')
y
```
